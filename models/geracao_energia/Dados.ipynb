```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho absoluto ou relativo corrigido
solar = pd.read_excel(r"C:\Users\Felipe Oliveira\Desktop\CIÊNCIA DE DADOS\SÉRIES TEMPORAIS\PROJETO DA DISCIPLINA\data\solar france.xlsx")

# Converter a coluna de data/hora
solar['Date and Hour'] = pd.to_datetime(solar['Date and Hour'])

# Filtrar apenas os 7 primeiros dias
inicio = solar['Date and Hour'].min()
fim = inicio + pd.Timedelta(days=7)
solar_7dias = solar[(solar['Date and Hour'] >= inicio) & (solar['Date and Hour'] < fim)]

# Plotar gráfico
plt.figure(figsize=(12,5))
plt.plot(solar_7dias['Date and Hour'], solar_7dias['valor'], linewidth=1.5)
plt.title('Geração de Energia Solar - Primeiros 7 dias')
plt.xlabel('Data e Hora')
plt.ylabel('Geração (valor)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
